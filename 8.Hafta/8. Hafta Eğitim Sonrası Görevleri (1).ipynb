# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde:  
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
# VKI hesaplama fonksiyonu
def calculate_BMI(weight, height):
    return weight / ((height/100)**2)

# Verileri kaydetmek için dosyayı açalım
with open("veriler.txt", "a") as f:

    # Kullanıcıdan kaç veri gireceğini alalım
    n = int(input("Kaç veri gireceksiniz? "))

    # Her veri için kullanıcıdan bilgileri alalım ve VKI hesaplayalım
    for i in range(n):
        name = input("İsim ve Soyisim: ")
        age = int(input("Yaş: "))
        height = float(input("Boy (cm): "))
        weight = float(input("Kilo (kg): "))
        bmi = calculate_BMI(weight, height)

        # VKI sınıfını belirleyelim
        if bmi < 18.5:
            vki_sinifi = "Zayıf"
        elif bmi < 25:
            vki_sinifi = "Normal kilolu"
        elif bmi < 30:
            vki_sinifi = "Fazla kilolu"
        elif bmi < 40:
            vki_sinifi = "Obez"
        else:
            vki_sinifi = "İleri derecede obez (morbid obez)"

        # Bilgileri ekrana yazdıralım
        print(f"\nİsim ve Soyisim: {name}")
        print(f"Yaş: {age}")
        print(f"Boy (cm): {height}")
        print(f"Kilo (kg): {weight}")
        print(f"Vücut Kitle İndeksi: {bmi:.2f}")
        print(f"VKİ Sınıfı: {vki_sinifi}")

        # Verileri dosyaya kaydedelim
        f.write(f"{name},{age},{height},{weight},{bmi:.2f},{vki_sinifi}\n")

```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import datetime

class Smoothie:
    def __init__(self, name, price, ingredients):
        self.name = name
        self.price = price
        self.ingredients = ingredients
    
    def __str__(self):
        return f"{self.name}: {self.price} TL"
    

class SmoothieBank:
    def __init__(self):
        self.smoothies = []
        self.transactions = []
        
    def add_smoothie(self, name, price, ingredients):
        smoothie = Smoothie(name, price, ingredients)
        self.smoothies.append(smoothie)
        
    def display_menu(self):
        print("Smoothie Menüsü")
        print("---------------")
        for smoothie in self.smoothies:
            print(smoothie)
        
    def purchase_smoothie(self, smoothie_index):
        smoothie = self.smoothies[smoothie_index]
        print(f"{smoothie.name} siparişi alındı. Toplam tutar: {smoothie.price} TL")
        now = datetime.datetime.now()
        transaction = {"time": now, "smoothie": smoothie}
        self.transactions.append(transaction)
        
    def display_transactions(self):
        for transaction in self.transactions:
            time = transaction["time"].strftime("%d.%m.%Y %H:%M:%S")
            smoothie = transaction["smoothie"]
            print(f"{time} - {smoothie.name} - {smoothie.price} TL")
        

bank = SmoothieBank()
bank.add_smoothie("Green Smoothie", 10, ["Ispanak", "Muz", "Zencefil"])
bank.add_smoothie("Protein Smoothie", 15, ["Yulaf", "Süt", "Çilek", "Badem"])
bank.add_smoothie("Detox Smoothie", 12, ["Elma", "Turp", "Limon", "Tarçın"])

while True:
    print("YetBar Smoothie Bank")
    print("--------------------")
    print("1- Menüyü Görüntüle")
    print("2- Satın Al")
    print("3- Satış Geçmişi")
    print("0- Çıkış")
    choice = input("Seçiminiz: ")
    
    if choice == "1":
        bank.display_menu()
    elif choice == "2":
        bank.display_menu()
        smoothie_index = int(input("Hangi smoothie'yi satın alacaksınız? (0, 1, 2): "))
        bank.purchase_smoothie(smoothie_index)
    elif choice == "3":
        bank.display_transactions()
    elif choice == "0":
        break
    else:
        print("Geçersiz seçim, tekrar deneyin.")
```
